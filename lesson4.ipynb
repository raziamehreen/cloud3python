```python
# enter day no and show day.
day_no = input("Enter day no.")
if day_no == "1":
    print("the day is monday")
elif day_no == "2":
    print("The day is Tuesday")
elif day_no == "3":
    print("The day is Wednessday")
elif day_no == "4":
    print("The day is Thursday")
elif day_no == "5":
    print("The day is Friday")
elif day_no == "6":
    print("The day is Saturday")
elif day_no == "7":
    print("The day is Sunday")
else:
    print("Please use proper day no between 1-7")
```

```python
 # enter temperature and show weather condition.
temperature = int(input("enter temperature"))
if temperature > 40: # above 40
    print("very hot")
elif temperature > 30: # 31-40
    print("warm")
elif temperature > 15: # 16-30
    print("normal")
elif temperature > 5: # 6-15
    print("cool")
elif temperature > -4: #-3-5
    print("cold")
else: 
    print("very cold") # below -4
```
